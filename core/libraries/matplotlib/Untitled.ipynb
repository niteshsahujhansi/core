# Simplest Way of Learning **matplotlib**

```python
from matplotlib import pyplot as plt
import numpy as np
```

# plt.title()

```python
plt.title('demo chart')
plt.xlabel('this is x label')
plt.ylabel('this is y label')
```

# fontdict, loc - plt.title() parameter/argument

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title('demo chart', fontdict=font1, loc='left')
plt.xlabel('this is x label')
plt.ylabel('this is y label', fontdict=font2)
```

# plt.plot()

```python
x = [1, 2, 6, 8]
y = [3, 8, 1, 10]
```

```python
x, y
```

```python
plt.plot(x,y)
```

# plt.plot() parameter - marker, linestyle, color, linewidth, markersize, markeredgecolor, markerfacecolor

```python
plt.title('demo chart')
plt.xlabel('this is x label')
plt.ylabel('this is y label')
plt.plot(x,y, marker='o', linestyle='dashdot', color='green', linewidth=5.5, markersize=20,
         markeredgecolor='red', markerfacecolor='yellow')
```

# Multiple line

```python
x2 = [2, 4, 7, 9]
y2 = [4, 9, 1, 11]
```

```python
plt.plot(x,y, label = 'first', linewidth=2)
plt.plot(x2,y2, label = 'secnd', linewidth=4)

plt.legend()
```

# plt.grid() - axis, color, linestyle, linewidth

```python
plt.plot(x,y)
plt.grid(color='green', linestyle='--',linewidth=0.5)
```

# plt.subplot(row, col, sr.no. of plot)

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

x2 = [2, 3, 7, 9]
y2 = [4, 9, 2, 11]

plt.subplot(1, 2, 2)
plt.plot(x2, y2)
```

# plt.scatter()

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x2, y2, color='green')
```

# Color, cmap, plt.colorbar(), size, alpha

```python
# colors = np.array(["red","green","blue","yellow","pink","black","orange","purple",
#                    "beige","brown","gray","cyan","magenta"])

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()
```

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)
```

# plt.pie()

```python
x = np.array([35, 25, 25, 15])
plt.pie(x)
plt.show()
```

# labels, explode, shadow, colors

# legend, title, loc, bbox_to_anchor

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

myexplode = [0.1, 0, 0, 0]

mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(x, labels = mylabels, explode=myexplode, shadow = True, colors=mycolors, autopct='%1.2f%%')
plt.legend(title = "Four Fruits:", bbox_to_anchor=(1.3, 0.8),  loc='right')
plt.show()
```
